Frequency Over Voltage

```python
import numpy as np
import pandas as pd
from numpy import *
import matplotlib.pyplot as plt

freq = array([1869.32, 1672.87, 1446.74, 1181.81, 834.44, 1.000])
voltage = array([-0.110, -0.088, -0.066, -0.044, -0.022, 0])

fig, ax = plt.subplots()

print ('Bayo-Adegoke Iyanuoluwa.')

plt.grid()
plt.xlabel('Voltage(V)', fontsize=10)
plt.ylabel('Frequency(GHz)',fontsize=10)
plt.title('Frequency(GHz) versus Voltage(V).',fontsize=15, color = "black")
plt.yticks(color='blue')
plt.xticks(color='blue')

line, = ax.plot(voltage,freq, color="blue")
f = np.linspace(0,10,100)

plt.scatter(voltage,freq, color ='black')

plt.show()
```
